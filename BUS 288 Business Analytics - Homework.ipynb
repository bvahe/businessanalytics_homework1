2. Fit a simple linear regressions (use LinearRegression from Scikit-learn), where Limit serves as a target variable and Income as feature (explanatory variable). Calculate and display intercept, coefficient and R^2.


```python
X = np.array(credit["Income"]).reshape(-1,1)
y = credit['Limit']

reg = LinearRegression()
reg.fit(X,y)
```

```python
print("coefficients = ", reg.coef_)
print("intercept = ",reg.intercept_)
print("prediction of the first observation = ", reg.predict(X)[0])
print("R^2 = ", reg.score(X, y, sample_weight=None))
```

```python
reg.score(X, y, sample_weight=None)
```

```python
fig, axes = plt.subplots(1, 1, figsize=(10, 5))
axes.scatter(X, y, alpha=0.3, label="ground truth")
axes.scatter(X, reg.predict(X), alpha=0.3, label="predictions")
axes.set(xlabel="Limit", ylabel="Income", title="Simple Linear Regression")
axes.legend()

plt.show()
```
